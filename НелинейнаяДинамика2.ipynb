```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def system(y, t, alpha):
    dxdt = [y[0] - y[0]*y[1]/(1+y[0])- alpha*y[0]*y[0], -y[1] + y[0]*y[1]/(1+y[0])]
    return dxdt
```

```python
param_values = [0.5, 1, 1.5]
colors = ['blue', 'green', 'red']
for i, alpha in enumerate(param_values):
  x0=np.linspace(-1,3, 10)
  x1=np.linspace(-2,2, 10)
  X0, X1 = np.meshgrid(x0,x1)
  shape1, shape2 = X1.shape
  dx0 = np.zeros(X0.shape)
  dx1 = np.zeros(X1.shape)
  plt.figure(figsize=(6, 6))
  for indexShape1 in range(shape1):
    for indexShape2 in range(shape2):
      dxdtAtX = system([X0[indexShape1, indexShape2], X1[indexShape1, indexShape2]], 0, alpha)
      dx0[indexShape1, indexShape2] = dxdtAtX[0]
      dx1[indexShape1, indexShape2] = dxdtAtX[1]
  plt.quiver(X0, X1, dx0, dx1, label=f'alpha = {alpha}', color=colors[i])
  plt.scatter(0, 0, label='Точка 0;0')
  plt.scatter(1/alpha, 0, label=f'Точка {1/alpha};0')
  plt.xlabel('x')
  plt.ylabel('y')
  plt.title(f'Phase Portrait for alpha = {alpha}')
  plt.legend()
  plt.grid(True)
  plt.show()
```
