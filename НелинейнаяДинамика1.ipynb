```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Параметры гармонического осциллятора
m = 1.0  # масса
k = 1.0  # жесткость
omega = np.sqrt(k/m)  # угловая частота

# Временной массив
t = np.linspace(0, 10, 1000)  # 0 до 10 секунд, 1000 точек

# Начальные условия
x0 = 1.0  # начальная позиция
v0 = 0.0  # начальная скорость

# Решение уравнений движения
# Позиция (x)
x = x0 * np.cos(omega * t)
# Скорость (v)
v = -x0 * omega * np.sin(omega * t)

# Создание 3D графика
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Построение линии для 3D фазового портрета
ax.plot(x, v, t, label='Фазовый портрет', color='#01008B')

# Настройка меток осей
ax.set_xlabel('Положение (x)')
ax.set_ylabel('Скорость (dx/dt)')
ax.set_zlabel('Время (t)')
ax.set_title('3D фазовый портрет гармонического осциллятора')

# Показать легенду
ax.legend()

# Показ графика
plt.show()
```

```python
from scipy.integrate import odeint

# Параметры осциллятора
m = 1.0  # масса
k = 1.0  # жесткость
c = 0.1  # коэффициент трения

# Уравнение движения
def model(state, t):
    x, v = state
    dxdt = v
    dvdt = - (c/m) * v - (k/m) * x
    return [dxdt, dvdt]

# Временной массив
t = np.linspace(0, 20, 1000)

# Начальные условия
initial_conditions = [1.0, 0.0]  # [начальная позиция, начальная скорость]

# Решение системы уравнений
solution = odeint(model, initial_conditions, t)
x = solution[:, 0]  # Позиция
v = solution[:, 1]  # Скорость

# Создание 3D графика
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Построение 3D графика
ax.plot(x, v, t, label='Фазовый портрет', color='green')

# Настройка меток осей
ax.set_xlabel('Положение (x)')
ax.set_ylabel('Скорость (dx/dt)')
ax.set_zlabel('Время (t)')
ax.set_title('3D фазовый портрет гармонического осциллятора с трением')

# Показать легенду
ax.legend()

# Показ графика
plt.show()
```
